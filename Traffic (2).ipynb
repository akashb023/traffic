Exploratory Data Analysis (EDA)

```python

from scipy import stats
df = pd.read_csv('traffic.csv')
df['Fine_Amount'] = pd.to_numeric(df['Fine_Amount'], errors='coerce')
df = df.dropna(subset=['Fine_Amount'])
mean = df['Fine_Amount'].mean()
median = df['Fine_Amount'].median()
mode = df['Fine_Amount'].mode().values[0] if not df['Fine_Amount'].mode().empty else None
std_dev = df['Fine_Amount'].std()
variance = df['Fine_Amount'].var()
range_ = df['Fine_Amount'].max() - df['Fine_Amount'].min()
df['Z_Score'] = stats.zscore(df['Fine_Amount'])
print(f"Mean: ₹{mean:.2f}")
print(f"Median: ₹{median}")
print(f"Mode: ₹{mode}")
print(f"Standard Deviation: ₹{std_dev:.2f}")
print(f"Variance: ₹{variance:.2f}")
print(f"Range: ₹{range_:.2f}")
print("\nFirst 5 Z-Scores:\n", df[['Fine_Amount', 'Z_Score']].head())
df.to_csv("fine_with_zscores.csv", index=False)
```

Statistical Analysis

```python
import pandas as pd
df = pd.read_csv("traffic.csv")
cols_to_convert = ['Fine_Amount', 'Penalty_Points', 'Speed_Limit', 'Recorded_Speed', 'Driver_Age', 'Alcohol_Level']
for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')
df_clean = df[cols_to_convert].dropna()
corr_matrix = df_clean.corr()
corr_pairs = corr_matrix.unstack().reset_index()
corr_pairs.columns = ['Feature_1', 'Feature_2', 'Correlation']
corr_pairs = corr_pairs[corr_pairs['Feature_1'] != corr_pairs['Feature_2']]
corr_pairs['Pairs'] = list(zip(corr_pairs.Feature_1, corr_pairs.Feature_2))
corr_pairs['Sorted_Pairs'] = corr_pairs['Pairs'].apply(lambda x: tuple(sorted(x)))
corr_pairs = corr_pairs.drop_duplicates(subset='Sorted_Pairs').drop(columns=['Pairs', 'Sorted_Pairs'])
corr_pairs['Abs_Correlation'] = corr_pairs['Correlation'].abs()
corr_pairs_sorted = corr_pairs.sort_values(by='Abs_Correlation', ascending=False)
print(" Top 5 Strongest Correlations:\n")
print(corr_pairs_sorted[['Feature_1', 'Feature_2', 'Correlation']].head(5).to_string(index=False))
```

Clustering

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
df = pd.read_csv("t.csv")
df['Fine_Paid'] = df['Fine_Paid'].map({'Yes': 1, 'No': 0})
df['Previous_Violations'] = pd.to_numeric(df['Previous_Violations'], errors='coerce')
df['Fine_Amount'] = pd.to_numeric(df['Fine_Amount'], errors='coerce')
df['Penalty_Points'] = pd.to_numeric(df['Penalty_Points'], errors='coerce')
df_encoded = pd.get_dummies(df[['Payment_Method']], prefix='Pay')
features = pd.concat([
    df[['Fine_Amount', 'Previous_Violations', 'Penalty_Points', 'Fine_Paid']],
    df_encoded
], axis=1).dropna()
scaler = StandardScaler()
scaled = scaler.fit_transform(features)
kmeans = KMeans(n_clusters=4, random_state=42)
features['Customer_Segment'] = kmeans.fit_predict(scaled)
segment_summary = features.groupby('Customer_Segment').mean().round(2)
print("Customer Segment Summary (based on behavioral clustering):\n")
print(segment_summary)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data
df = pd.read_csv("traffic.csv")

# Convert Fine_Amount to numeric (in case it's not)
df['Fine_Amount'] = pd.to_numeric(df['Fine_Amount'], errors='coerce')

# Drop missing values for Fine_Amount and Location
df = df.dropna(subset=['Fine_Amount', 'Location'])

# Group by Location and calculate mean fine amount
mean_fines = df.groupby('Location')['Fine_Amount'].mean().reset_index()

# KMeans Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
mean_fines['Cluster'] = kmeans.fit_predict(mean_fines[['Fine_Amount']])

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=mean_fines, x='Location', y='Fine_Amount', hue='Cluster', palette='Set2', s=100)

plt.title('Clustering of Locations by Mean Fine Amount')
plt.xlabel('Location')
plt.ylabel('Mean Fine Amount')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Simple Linear Regression :

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("traffic.csv")  
df['Fine_Amount'] = pd.to_numeric(df['Fine_Amount'], errors='coerce')
df['Speed_Limit'] = pd.to_numeric(df['Speed_Limit'], errors='coerce')
df['Recorded_Speed'] = pd.to_numeric(df['Recorded_Speed'], errors='coerce')
df['Alcohol_Level'] = pd.to_numeric(df['Alcohol_Level'], errors='coerce')
df['Penalty_Points'] = pd.to_numeric(df['Penalty_Points'], errors='coerce')

df = df.dropna(subset=['Fine_Amount', 'Speed_Limit', 'Recorded_Speed', 'Alcohol_Level', 'Penalty_Points'])
X = df[['Speed_Limit', 'Recorded_Speed', 'Alcohol_Level', 'Penalty_Points']]
X = sm.add_constant(X)  
y = df['Fine_Amount']
model = sm.OLS(y, X).fit()
print(model.summary())
df['Predicted_Fine'] = model.predict(X)
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Fine_Amount'], y=df['Predicted_Fine'], color='darkorange', s=60)
plt.plot([df['Fine_Amount'].min(), df['Fine_Amount'].max()],
         [df['Fine_Amount'].min(), df['Fine_Amount'].max()],
         'r--')
plt.xlabel("Actual Fine Amount")
plt.ylabel("Predicted Fine Amount")
plt.title("Actual vs Predicted Fine Amount")
plt.grid(True)
plt.tight_layout()
plt.show()
```

Evaluate models using R², standard error of estimate.

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np

# Load your dataset
df = pd.read_csv("traffic.csv")  # Replace with your actual file path

# Convert necessary columns to numeric
df['Fine_Amount'] = pd.to_numeric(df['Fine_Amount'], errors='coerce')
df['Speed_Limit'] = pd.to_numeric(df['Speed_Limit'], errors='coerce')
df['Recorded_Speed'] = pd.to_numeric(df['Recorded_Speed'], errors='coerce')
df['Alcohol_Level'] = pd.to_numeric(df['Alcohol_Level'], errors='coerce')
df['Penalty_Points'] = pd.to_numeric(df['Penalty_Points'], errors='coerce')

# Drop missing values
df = df.dropna(subset=['Fine_Amount', 'Speed_Limit', 'Recorded_Speed', 'Alcohol_Level', 'Penalty_Points'])

# Define features (X) and target (y)
X = df[['Speed_Limit', 'Recorded_Speed', 'Alcohol_Level', 'Penalty_Points']]
X = sm.add_constant(X)
y = df['Fine_Amount']

# Fit the multiple regression model
model = sm.OLS(y, X).fit()

# Print R²
print(f"R² (Coefficient of Determination): {model.rsquared:.4f}")

# Calculate Standard Error of Estimate (SEE)
residuals = model.resid
n = len(y)
k = X.shape[1] - 1  # Exclude intercept
SEE = np.sqrt(sum(residuals**2) / (n - k - 1))
print(f"Standard Error of Estimate (SEE): {SEE:.4f}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("traffic.csv")  
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df = df.sort_values(by=['Driver_Gender', 'Date'])  
repeat_df = df.groupby('Driver_Gender').filter(lambda x: len(x) >= 2)
repeat_df['Violation_Order'] = repeat_df.groupby('Driver_Gender').cumcount() + 1
pivot_df = repeat_df[repeat_df['Violation_Order'] <= 2].pivot(index='Driver_Gender', columns='Violation_Order', values='Fine_Amount')
pivot_df.columns = ['First_Fine', 'Second_Fine']
pivot_df = pivot_df.dropna()
plt.figure(figsize=(8, 5))
sns.scatterplot(data=pivot_df, x='First_Fine', y='Second_Fine', alpha=0.6)
sns.lineplot(x=pivot_df['First_Fine'], y=pivot_df['First_Fine'], color='red', label='y = x')
plt.title('Regression to the Mean: First vs Second Fine')
plt.xlabel('First Fine Amount')
plt.ylabel('Second Fine Amount')
plt.grid(True)
plt.legend()
plt.show()
```
